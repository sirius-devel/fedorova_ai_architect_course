```python
################################TASK 1
A = np.array([[1, 2, 1], [-2, 1, 1]])
B = np.array([[3, 1], [-1, 0], [2, 4]])
C = np.array([[-2, 2], [1, -1], [1, 2]])
R = A + 2*np.transpose(B)
print(R)
R = B+ 2*C
print(R)
```

```python
##########################TASK3
A = np.arange(0, 1.5, 0.1)
A = A.reshape(5, 3)
print(A)
B = np.arange(1, 7, 1)
B = B.reshape(3, 2)
print(B)
R = np.dot(A, B)
print(R)
```
