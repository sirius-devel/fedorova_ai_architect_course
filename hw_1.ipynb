```python
def matrix_product(first_matrix, second_matrix):
  row_count1 = len(first_matrix)
  row_count2 = len(second_matrix)
  col_count2 = len(second_matrix)
  result_matrix = [[0 for i in range(col_count2)] for i in range(row_count1)]
  for i in range(row_count1):
    for j in range(col_count2):
      for k in range(row_count2):
        result_matrix[i][j] += first_matrix[i][k] * second_matrix[k][j]
  return result_matrix


first_matrix = [[6, -8, 3]]
second_matrix = [[1, 1, 4]]
print(matrix_product(first_matrix, second_matrix))

first_matrix = [[-3, -7], [-10, 6], [0, -6], [-1, -3]]
second_matrix = [[-8, 9], [4, -2]]
print(matrix_product(first_matrix, second_matrix))
```
